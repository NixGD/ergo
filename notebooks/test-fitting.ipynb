```python
from ergo.distributions import PointDensity
from ergo.conditions import IntervalCondition, MaxEntropyCondition
from ergo.scale import Scale
import jax
```

```python
with jax.disable_jit():
    d = PointDensity.from_conditions([
        IntervalCondition(min=30,max=60,p=.6,weight=20),
        MaxEntropyCondition(weight=.01)
    ], scale=Scale(0,100))
```
