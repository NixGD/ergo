```python
import ergo
```

```python
from ergo.platforms.metaculus.question import LinearQuestion, LinearDateQuestion
```

```python
metaculus = ergo.Metaculus(username="oughtpublic", password="123456", api_domain="pandemic")
```

```python
open_qs = metaculus.get_questions(question_status="open", player_status="any", pages=9999)
```

```python
actual_qs = [q for q in open_qs if q is not None]
```

```python
linear_qs = [q for q in actual_qs if isinstance(q, LinearQuestion) and not isinstance(q, LinearDateQuestion)]
```

```python
linear_qs[0].to_dataframe(linear_qs)
```

```python
open_qs = [q for q in linear_qs if q.low_open or q.high_open]
```
