```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
```

```python
from ergo.platforms.metaculus.question import LinearQuestion, LinearDateQuestion
```

```python
metaculus = ergo.Metaculus(username="oughtpublic", password="123456")
```

```python
open_qs = metaculus.get_questions(question_status="open", pages=9999)
```

```python
actual_qs = [q for q in open_qs if q is not None]
```

```python
linear_qs = [q for q in actual_qs if isinstance(q, LinearQuestion) and not isinstance(q, LinearDateQuestion)]
```

```python
linear_qs[0].to_dataframe(linear_qs)
```

```python
open_qs = [q for q in linear_qs if q.low_open or q.high_open]
```

```python
open_qs[0].to_dataframe(open_qs)
```

```python
open_qs_df = open_qs[0].to_dataframe(open_qs)
```

```python
p_outside = [(q.latest_community_percentiles["low"] + (1 - q.latest_community_percentiles["high"])) for q in open_qs]
open_qs_df["p_outside"] = p_outside
low_open = [q.low_open for q in open_qs]
open_qs_df["low_open"] = low_open
high_open = [q.high_open for q in open_qs]
open_qs_df["high_open"] = high_open
```

```python
open_qs_df.sort_values("p_outside", ascending=False)
```
